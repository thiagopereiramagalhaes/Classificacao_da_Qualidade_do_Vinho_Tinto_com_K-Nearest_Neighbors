#### Apresente a Curva ROC

Dado que a Curva ROC é uma métrica para valores binários, vamos fazer o cálculo de uma classe e comparar com as demais classes.

Referência:
https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html

É preciso binarizar as variáveis independentes, calcular as probabilidades de cada predição para target e, por fim calcular a curva ROC.

```python
from sklearn.preprocessing import LabelBinarizer
from sklearn.metrics import roc_curve, auc
from sklearn.metrics import RocCurveDisplay
```

Binarização da variável dependente, quality.

```python
label_binarizer = LabelBinarizer().fit(df['quality'])
```

```python
y_onehot_test = label_binarizer.transform(y_test)
y_onehot_test.shape  # (n_samples, n_classes)
```

Selecionando a qualidade de número **5** pois foi o maior valor obtido pelo modelo.

```python
class_of_interest = 5
class_id = np.flatnonzero(label_binarizer.classes_ == class_of_interest)[0]
class_id
```

Calculando as probabilidades dos dados de teste para fazer o cálculo da curva ROC.

```python
y_scores = knn.predict_proba(X_test_scaled)
```

```python
def plot_curva_ROC_one_vs_rest(y_test, y_scores):

  RocCurveDisplay.from_predictions(
      y_onehot_test[:, class_id],
      y_scores[:, class_id],
      name=f"{class_of_interest} vs as demais classes",
  )
  plt.plot([0, 1], [0, 1], "k--", label="valor de threashold (AUC = 0.5)")
  plt.axis("square")
  plt.xlabel("Taxa de Falso Positivo")
  plt.ylabel("Taxa de Verdadeira Positivo")
  plt.title("One-vs-Resto curva ROC")
  plt.legend()
  plt.show()
```

```python
plot_curva_ROC_one_vs_rest(y_test, y_scores)
```

# Escolhendo o valor K

Vamos continuar usando o método do cotovelo para escolher um bom valor K!

Crie um loop for que treine vários modelos KNN com valores k diferentes e, em seguida, mantenha um registro do error_rate para cada um desses modelos com uma lista. Consulte o notebook se você estiver confuso nesta etapa.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

def train_modelo(X_train_scaled, X_test_scaled, y_train, k=1):
  print(f"Treinando o k-NN com k={k}")
  #faça a instância de um novo modelo, treine esse modelo e faça as predições para os dados de teste
  knn_model = KNeighborsClassifier(n_neighbors=k)
  knn_model.fit(X_train_scaled, y_train)
  pred_i = knn_model.predict(X_test_scaled)
  return pred_i
```

```python
error_rate = []

for i in range(1,41):
  pred_i = train_modelo(X_train_scaled, X_test_scaled, y_train, i)
  #calcule e armazene os valores de erro para cada modelo treinado

  error = 1 - accuracy_score(y_test, pred_i)
  error_rate.append(error)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,41),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
plt.show()
```

### Conclusão do Projeto KNN - Qualidade do Vinho

Neste projeto, o objetivo foi aplicar o algoritmo K-Nearest Neighbors (KNN) para classificar a qualidade de vinhos tintos, utilizando a base de dados Wine da UCI. A abordagem seguiu as seguintes etapas:

1.  **Análise Descritiva dos Dados (EDA):**
    *   **Carregamento e Exploração Inicial:** Os dados foram carregados de um arquivo CSV e explorados utilizando `df.head()`, `df.info()` e `df.describe()` para entender a estrutura, tipos de dados e estatísticas descritivas das variáveis. A análise `df.info()` confirmou a ausência de valores nulos, simplificando a etapa de pré-processamento. `df.describe()` forneceu um panorama das distribuições das features.
    *   **Análise de Valores Ausentes:** Utilizou-se `missingno.matrix()` para visualizar graficamente a completude dos dados, confirmando que não havia valores faltantes, o que evitou a necessidade de estratégias de imputação.
    *   **Distribuição dos Dados:** Histogramas foram gerados para cada feature (`df.hist()`), revelando a distribuição de cada variável e indicando a presença de algumas distribuições assimétricas que poderiam beneficiar a normalização.
    *   **Análise de Correlação:** Um mapa de calor (`sns.heatmap(df.corr())`) foi utilizado para visualizar as correlações entre as features, auxiliando na identificação de relações lineares e possíveis recursos importantes para o modelo.

2.  **Divisão Treino-Teste:**
    *   Os dados foram divididos em conjuntos de treino e teste (`X_train`, `X_test`, `y_train`, `y_test`) utilizando `train_test_split` com uma proporção de 75% para treino e 25% para teste, e `random_state=111` para garantir reprodutibilidade. A variável alvo (`quality`) foi separada das features.

3.  **Normalização das Variáveis:**
    *   Para garantir que todas as features contribuíssem igualmente para o cálculo da distância no algoritmo KNN, os dados independentes (`X_train` e `X_test`) foram normalizados usando `StandardScaler`. Isso padroniza as features para terem média zero e desvio padrão um. A normalização foi verificada através da inspeção do `pd.DataFrame(X_train_scaled).head()` e da aplicação do `scaler.inverse_transform` para confirmar a reversibilidade.

4.  **Treinamento e Avaliação do Modelo KNN (K=1):**
    *   Um primeiro modelo KNN foi instanciado com `n_neighbors=1` e treinado com os dados escalados. A previsão foi feita no conjunto de teste e avaliada com um `classification_report` e uma matriz de confusão. A acurácia inicial foi de 62%, mas o relatório destacou um desempenho fraco para as classes de vinho com menos amostras (qualidade 3, 4 e 8), que são classes minoritárias.
    *   A curva ROC 'One-vs-Rest' foi gerada para a classe de interesse (qualidade 5), mostrando o desempenho do classificador binário para distinguir esta classe das demais.

5.  **Otimização do K (Método do Cotovelo):**
    *   Para encontrar um valor ideal para `K`, foi implementado um loop que treinava modelos KNN para `K` variando de 1 a 40. Para cada `K`, a taxa de erro (`error_rate`) foi calculada e armazenada. O gráfico da 'Taxa de Erro vs. Valor de K' (método do cotovelo) foi plotado para visualizar como a taxa de erro muda com diferentes valores de `K`.
    *   Observando o gráfico, a taxa de erro inicial é alta para `K=1` e diminui consideravelmente, estabilizando e atingindo um mínimo em torno de `K=20` a `K=30`. Com base nessa análise, o valor de **K=23** foi escolhido como um bom equilíbrio para o modelo.

6.  **Re-treinamento e Reavaliação com K Otimizado:**
    *   O modelo KNN foi re-treinado com o valor de `n_neighbors=23`. O `classification_report` e a matriz de confusão foram gerados novamente, e a curva ROC também foi plotada. A expectativa é que o desempenho geral do modelo tenha melhorado, especialmente em termos de estabilidade e generalização, comparado ao modelo com `K=1`.
